```python
def out_det(x):
    outliers=[]; 
    max=3
    m=np.mean(x)
    s=statistics.stdev(x) 
 
    for y in x:
      z= (y-m)/s 
      if np.abs(z) > max:
         outliers.append(y)
    return outliers 
```

```python
print("BMI=" ,out_det(insurance.bmi))
print("Age=" , out_det(insurance.age))
print("charges=" ,out_det(insurance.charges))
```

# 4. Answer the following questions with statistical evidence 

### a. Do charges of people who smoke differ significantly from the people who don't?


```python
ms=insurance.charges.sample(n=200)[insurance['smoker']=='yes'].mean()
stdms=statistics.stdev(insurance.charges.sample(n=200)[insurance['smoker']=='yes']) 
mnons=insurance.charges.sample(n=200)[insurance['smoker']=='no'].mean() 
stdnons=statistics.stdev(insurance.charges.sample(n=200)[insurance['smoker']=='no']) 
```

##### h0= ms=mnons (Smoking does not affect charges)<br>h1= sym=!snm (Smoking affects charges)<br>alpha=0.05

```python
Z=(ms-mnons)/(math.sqrt(((stdms*stdms)/200)+((stdnons*stdnons)/200)))
Z
```

#### Z score is not between the interval [-1.96 and +1.96], hence, Null (ho) hypothesis Rejected. <br> Smoking Affects charges

### b. Does bmi of males differ significantly from that of females?

```python
mm=insurance.bmi.sample(n=200)[insurance['sex']=='male'].mean()
mf=insurance.bmi.sample(n=200)[insurance['sex']=='female'].mean()
sm=statistics.stdev(insurance.charges.sample(n=200))
sf=statistics.stdev(insurance.charges.sample(n=200))
```

#### h0= meanmale=meanfemale (Bmi does not differs significantly) <br> h1= meanmale=!meanfemale (Bmi differs significantly) <br> alpha=0.10

```python
Z=(mm-mf)/(math.sqrt(((sm*sm)/200)+((sf*sf)/200)))
Z
```

#### Z score is between the interval [-1.96 and +1.96], hence, failed to reject Null (ho) hypothesis <br> Bmi does not differ significantly

### c. Is the proportion of smokers significantly different in different genders?

#### Given, n= 1338, male=number of male smokers, female= number of female smokers, P(alpha)= 0.5
#### H0= Assume both the proportions are same i.e, P=0.5
#### H1= Assume both the proportions are not same i.e, P≠0.5
#### Z= p-P/√PQ/n

```python
male=insurance.smoker[insurance['sex']=='male'].count()
female=insurance.smoker[insurance['sex']=='female'].count()
p=male/1338
P=0.5
Q=1-0.5
n=1338
z=(p-P)/math.sqrt((P*Q)/n)
z
```

#### Z score is between the interval [-1.96 and +1.96], hence, failed to reject Null (ho) hypothesis <br> No,Proportions are not different
