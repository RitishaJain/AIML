```python
#As we visualised in the pairplot there are mix up of gaussians in the dataset.
#So we divide the data into clusters using k means clustering.
```

```python
from sklearn.cluster import KMeans
```

```python
cement = cement.loc[:,'cement':'strength']
cement_z = cement.apply(zscore)
```

```python
cluster_range = range( 3, 9 )   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(cement_z)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:9]
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
#AS we can see there is sharp elbow at 5 so we take the number of clusters to be 5.
```

```python
cluster = KMeans( n_clusters = 5, random_state = 2354 )
cluster.fit(cement_z)

prediction=cluster.predict(cement_z)
cement_z["GROUP"] = prediction    

cement_z_copy = cement_z.copy(deep = True)  
```

```python
#Visulisaing the outliers in every cluster using boxplot.
```

```python
import matplotlib.pylab as plt

cement_z.boxplot(by = 'GROUP',  layout=(2,9), figsize=(15, 10))
```

```python
#As we can see outliers are present in every single attribute so we try to replace it with the medians of that particular attribute.
```

```python
data = cement_z  
def replace(group):
    median, std = group.median(), group.std()  #Get the median and the standard deviation of every group 
    outliers = (group - median).abs() > 2*std # Subtract median from every member of each group. Take absolute values > 2std
    group[outliers] = group.median()       
    return group

data_corrected = (data.groupby('GROUP').transform(replace)) 
concat_data = data_corrected.join(pd.DataFrame(cement_z['GROUP']))
```

```python
print(pca.explained_variance_)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.bar(list(range(1,9)),pca.explained_variance_ratio_,alpha=0.5, align='center')
plt.ylabel('Variation explained')
plt.xlabel('eigen Value')
plt.show()
```

```python
plt.step(list(range(1,9)),np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Cum of variation explained')
plt.xlabel('eigen Value')
plt.show()
```

```python
#Checking the accuracy at 95% confidence level using KFold Cross Validation.
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
num_folds=80
seed=7

kfold=KFold(n_splits=num_folds, random_state=seed)
model= Ridge()
results=cross_val_score(model,X,y, cv=kfold)
print(results)
print("accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0,results.std()*100.0))
```
